## spiral order Given an m x n matrix, return all elements of the matrix in spiral order.

```python
import numpy as np
def spiralOrder(matrix): 
    matrix = np.array(matrix) #convert lists to array so we can use numpy. 
    sol = [] #solution to store answers. 
    while np.shape(matrix)[0] * np.shape(matrix)[1] > 0: #as long as we haven't scene every number. 
        sol = sol + matrix[0,:].tolist() #take first row. 
        matrix = np.delete(matrix,0,0) #delete it. 
        matrix = np.rot90(matrix) #rotate the remaining matrix so that we can take the next top row. $
    return sol
```
