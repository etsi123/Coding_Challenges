## Container with most water problem. Given n non-negative integers a1, a2, ..., an , where each represents a point at coordinate (i, ai). n vertical lines are drawn such that the two endpoints of the line i is at (i, ai) and (i, 0). Find two lines, which, together with the x-axis forms a container, such that the container contains the most water.

```python
from IPython.display import Image
Image(filename = 'maxWater.png', width=500, height=500)
```

```python
def maxWater_fast(input_list): 
    num_iterations=0 #count number of iterations
    num_elements = len(input_list)
    left_index,right_index = 0,num_elements-1
    maxWater = 0 #initialize amount of water
    while left_index < right_index: 
        num_iterations=num_iterations+1 
        x = right_index-left_index #width
        y = min(input_list[left_index],input_list[right_index]) #determine y
        maxWater = max(maxWater,x*y) #keep larger volume
        if input_list[left_index]>=input_list[right_index]: #change next index
            right_index = right_index-1 
        else: 
            left_index = left_index+1
    print('Num Iterations : ' + str(num_iterations))
    return maxWater
print(maxWater_fast([1,8,6,2,5,4,8,3,7]))
print(maxWater_fast([3,1,2,4,5]))
```

```python
def maxWater_slow(input_list):
    num_iterations = 0 
    num_elements = len(input_list)
    maxWater = 0
    for i in range(0,num_elements-1): 
        for j in range(i+1,num_elements):
            num_iterations+=1
            x = j-i
            y = min(input_list[i],input_list[j])
            maxWater = max(maxWater,x*y)
    print('Num Iterations : ' + str(num_iterations))
    return maxWater
print(maxWater_slow([1,8,6,2,5,4,8,3,7]))
print(maxWater_slow([3,1,2,4,5]))
print(maxWater_slow([1,5,4,3]))
```
