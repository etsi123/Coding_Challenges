## You are given two strings s and t. String t is generated by random shuffling string s and then add one more letter at a random position. Return the letter that was added to t.

```python
def findTheDifference(s,t): 
    for char in t: #loop over chars in string
        if char in s: #if char is present in s
            if s.count(char) != t.count(char): #make sure that its not an extra duplicate character
                print('There is an extra ' + str(char) + ' present.')
                return char                
        else: #character wasn't in string s, this is the extra character
            print(char + ' is the extra char present.')
            return char
        
findTheDifference("abcd","abecd")
findTheDifference("","a")
findTheDifference("a","aa")
findTheDifference("ae","aea")
```
