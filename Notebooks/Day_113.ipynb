## WallStreetBets (r/wallstreetbets, also known as WSB), is a subreddit where participants discuss stock and option trading. Find the most common words used in the title. 

```python
import numpy as np # linear algebra
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('reddit_wsb.csv')
df.head()
```

```python
titles = df['title'].apply(lambda x:str(x).split()) #split string
top = Counter([title for sublist in titles for title in sublist])
pd.DataFrame(top.most_common(20),columns = ['words','count'])
```
