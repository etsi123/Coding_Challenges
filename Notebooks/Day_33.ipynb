## Given an integer array nums of 2n integers, group these integers into n pairs (a1, b1), (a2, b2), ..., (an, bn) such that the sum of min(ai, bi) for all i is maximized. Return the maximized sum.

```python
from IPython.display import Image
Image(filename = 'day_33_explain.png', width=500, height=500)
```

```python
def arrayPairSum(input_list): 
    input_list.sort() #we want to maximize the sum, so let's sort so that we can pair biggest numbers with
                      #biggest numbers and smallest numbers with smallest numbers. 
    num_elements = len(input_list) 
    max_sum = 0 #keep track of sum
    for i in range(0,num_elements,2): #increment by 2 so that we take the smaller of the pair of numbers. 
        max_sum = max_sum + input_list[i] #add the smallest numbers in each ordered pair to sum. 
    return max_sum
```

```python
print(arrayPairSum([1,4,3,2]))
print(arrayPairSum([6,2,6,5,1,2]))
```
